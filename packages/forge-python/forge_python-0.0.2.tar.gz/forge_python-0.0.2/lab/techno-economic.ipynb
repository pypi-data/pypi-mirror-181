```python
import random
import requests
import json
from pandas import json_normalize
import time
import numpy as np


def sendSensorData(name, value):
    # sensor place holder
    sensor_name = name
    sensor_value = value
    sensor_id = 'device6ef0e2a6f88f481ea81e388d8be91a1c'

    # request parameters
    device_id = 'device6ef0e2a6f88f481ea81e388d8be91a1c'
    base_url = 'https://connect.forge-api.com'
    body = {
        "sensor_data": {
            'sensor_name': sensor_name,
            'sensor_value': sensor_value,
            'sensor_id': sensor_id,
        },
        "device_id": device_id,
    }
    headers = {'User-Agent': 'XY', 'Content-type': 'application/json'}
    response = requests.post(f"{base_url}/devices", json=body, headers=headers, params={
        "device_id": device_id,
    })
    data = response.json()

    json_normalize(data)
    return data


def createSamples(start_temperature, end_temperature, amount):
    arr = np.linspace(start_temperature, end_temperature, amount)
    def roundVal(val): return round(val, 1)
    applyall = np.vectorize(roundVal)(arr)
    return applyall


# simulation variables
start_temperature = 35
end_temperature = 37.5
temperature_samples = createSamples(start_temperature, end_temperature, 10)
```

```python
from colorama import init as colorama_init
from colorama import Fore
from colorama import Style

colorama_init()

print(f"This is {Fore.GREEN}color{Style.RESET_ALL}!")

class Forge: 
    class sensor:
        def __init__(self, name=''):
            self.name = name
            self.samples = None
            self.index = 0
            self.device_id = None
            self.sensor_name = None
            self.sensor_value = None

        def authenticate(self, device_id):
            # authenticate
            self.device_id = device_id
            return self
            
        def send(self, dict={}):
            self.sensor_name = list(dict.keys())[0]
            self.sensor_value = dict[self.sensor_name]

            # wait 1 second
            time.sleep(1)

            # @todo make this a special color
            print(f"[send] {self.sensor_name} value of {self.sensor_value} to remote")
            if self.device_id is not None:
                sendSensorData(self.sensor_name, self.sensor_value)
            else:
                print("Device id not authenticated or present")





```

```python
# send sensor event
sensor = Forge.sensor()
sensor.authenticate('device6ef0e2a6f88f481ea81e388d8be91a1c')
sensor.send({'temperature': 35.5})

for value in createSamples(start_temperature=35.5, end_temperature=37.5, amount=50):
    sensor.send({'temperature': value})
```
